```python
# Read csv
import pandas as pd
import os
df = pd.read_csv( os.getcwd() + '/2019_10210010.csv', encoding="shift-jis" )
```

```python
# Add new column YYYYMMDD-HH:00:00
df = pd.concat([df, df["日付"].str.split("/", expand=True)], axis=1)
df.rename(columns={0: "YYYY", 1: "MM", 2: "DD"}, inplace=True)
df["日時"] = df["YYYY"] + "-" + df["MM"] + "-" + df["DD"] + " " + (df["時"]-1).astype(str) + ":00:00"
df["日時"] = pd.to_datetime(df["日時"])
df = df.set_index("日時")
```

```python
# Check dataframe
df
```

```python
# Average value
df.mean()
```

```python
# Medium value
df.median()
```

```python
# Variance 
df.var()
```

```python
# Standard deviation
df.std()
```

```python
# Plot
import matplotlib.pyplot as plt
plt.scatter(df.index, df["PM2.5(ug/m3)"])
plt.show()
```

```python
# Make daily data
daily = pd.DataFrame(df["PM2.5(ug/m3)"].resample("D").mean())
```

```python
# Check daily data
daily
```

```python
# Plot daily data
plt.scatter(daily.index, daily["PM2.5(ug/m3)"])
plt.show()
```

```python
# Read and format weather data
weather = pd.read_csv(os.getcwd() + '/rainfall.csv', encoding="shift-jis", dtype=None)
weather["年月日"] = pd.to_datetime(weather["年月日"])
weather = weather.set_index("年月日")
```

```python
# Check weather data
weather
```

```python
# Join the data
data = daily.join(weather)
```

```python
# Check data
data
```

```python
# rainfall and PM2.5
plt.scatter(data.index, data["PM2.5(ug/m3)"], c='blue', label='PM2.5(ug/m3)')
plt.scatter(data.index, data["降水量(mm)"], c='red', label='Precipitation')
plt.legend(loc='upper right')
plt.title('Precipitation and PM2.5')
plt.xlabel('Date')
plt.ylabel('Precipitation')
plt.show()
```

```python
# Correlation
data.corr()
```

```python
# read and format weather data
temp = pd.read_csv(os.getcwd() + '/temp.csv', encoding="shift-jis",  dtype=None)
temp["年月日"] = pd.to_datetime(temp["年月日"])
temp = temp.set_index("年月日")
```

```python
# Check temperature data
temp
```

```python
# Join the data
data = daily.join(temp)
```

```python
# Check data
data
```

```python
# Temperature and PM2.5
plt.scatter(data.index, data["PM2.5(ug/m3)"], c='blue', label='PM2.5(ug/m3)')
plt.scatter(data.index, data["気温(℃)"], c='orange', label='Temperature')
plt.legend(loc='upper right')
plt.title('Temperature and PM2.5')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

```python
# Correlation
data.corr()
```
